## 模擬版資料做 SVM

```python
x = np.array([[6,6],[10,8],[3,7],[8,7]])
```

```python
x
```

```python
x[:,0]
## [每一列都要:,要的是第0行]
```

### Step 1.打開函數學習機

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma='auto')
```

```python
clf.fit(x,y)
```

### Step 3.預測

```python
y_predict = clf.predict(x)
```

```python
y_predict
```

## 鳶尾花數據庫

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris.DESCR)
```

```python
x = iris.data
y = iris.target
```

```python
x[:5]
```

```python
y[:5]
```

```python
X = x[:,:2]
```

```python
Y = y
```

```python
len(X)
```

### 區分訓練、測試資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, cmap='Set3')
```

### Step 1.開一台函數學習機

```python
clf = SVC(gamma='auto')
```

### Step 2.訓練

```python
clf.fit(x_train, y_train)
```

### Step 3.預測

```python
y_predict = clf.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
plt.scatter(x_test[:,0], x_test[:,1], c = y_predict-y_test)
```

```python
x0 = np.arange(3.8, 8.2, 0.02)
y0 = np.arange(1.8, 4.7, 0.02)

xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(),ym.ravel()]
z = clf.predict(P)

Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, cmap='Set3', alpha=0.3)
plt.scatter(x_test[:,0], x_test[:,1], cmap='Set3',c=y_test)
```
