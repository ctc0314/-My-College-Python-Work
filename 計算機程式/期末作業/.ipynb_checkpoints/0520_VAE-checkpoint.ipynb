```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, IntSlider, FloatSlider

import tensorflow as tf
import tensorflow.keras.backend as K
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, Lambda, concatenate
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras import metrics
from tensorflow.keras import losses
from tensorflow.keras import datasets
from tensorflow.keras.utils import to_categorical
```

### 資料整理

```python
(X_train, y_train0), (X_test, y_test0) = datasets.mnist.load_data()
```

```python
X_train = X_train.reshape(-1, 28*28)
X_test = X_test.reshape(-1, 28*28)
```

```python
X_train = X_train / X_train.max()
X_test = X_test / X_test.max()
```

```python
y_train = to_categorical(y_train0, 10)
y_test = to_categorical(y_test0, 10)
```

```python
x = Input(shape=(784,))
```

### VAE建立

```python
enc_1 = Dense(100, activation='sigmoid') 
enc_2 = Dense(235, activation='relu') ##調整參數
enc_3 = Dense(235, activation='sigmoid') ##增加第三層encoding層

enc_mean = Dense(2)
enc_log_var = Dense(2)

dec_3 = Dense(150, activation='sigmoid') ##調整參數
dec_2 = Dense(350, activation='sigmoid') ##調整參數
dec_1 = Dense(784, activation='sigmoid') ##增加第三層decoding層
```

```python
x = Input(shape=(784,))
enc_x = enc_3(enc_2(enc_1(x)))

z_mean = enc_mean(enc_x)
z_log_var = enc_log_var(enc_x)
```

```python
def sampling(args):
    z_mean, z_log_var = args
    batch = K.shape(z_mean)[0]
    dim = K.int_shape(z_mean)[1]
    # by default, random_normal has mean = 0 and std = 1.0
    epsilon = K.random_normal(shape=(batch, dim))
    return z_mean + K.exp(0.5 * z_log_var) * epsilon    
```

```python
# Sampling function wrapped as a Keras layer
z = Lambda(sampling, output_shape=(2,))([z_mean, z_log_var])
```

```python
z_input = Input(shape=(2,))
x_hat = dec_1(dec_2(dec_3(z_input)))
```

```python
x_hat = dec_1(dec_2(dec_3(z)))

VAE = Model(x, x_hat)
VAE.summary()
```

### Loss 函數

```python
# reconstruction_loss = tf.keras.losses.mse(inputs, outputs)
reconstruction_loss = 784 * losses.binary_crossentropy(x, x_hat)

kl_loss = 1 + z_log_var - K.square(z_mean) - K.exp(z_log_var)
kl_loss = K.sum(kl_loss, axis=-1)
kl_loss *= -0.7 ##-0.7
vae_loss = K.mean(reconstruction_loss + kl_loss)

VAE.add_loss(vae_loss)
```

```python
VAE.compile(optimizer=Adam())
```

```python
VAE.fit(X_train, 
        batch_size=1024, ##1024 
        epochs=8) ##訓練8次
```

### VAE 的視覺化

```python
VAE_Encoder = Model(x, z_mean)

VAE_Encoder.summary()
```

```python
VAE_Decoder = Model(z_input, dec_1(dec_2(dec_3(z_input))))

VAE_Decoder.summary()
```

```python
def normalized(x):
    x -= x.min()
    x /= x.max()
    return x
```

```python
n = 13 ##13
digit_size = 28
figure = np.zeros((digit_size * n, digit_size * n))
```

```python
grid_x_vae = np.linspace(-4+0.05, 4-0.05, n)
grid_y_vae = np.linspace(-4+0.05, 4-0.05, n)
VAE_figure = np.zeros((digit_size * n, digit_size * n))
```

```python
for i, yi in enumerate(grid_x_vae):
    for j, xi in enumerate(grid_y_vae):
        z_sample = np.array([[xi, yi]])
        x_decoded = VAE_Decoder.predict(z_sample)
        digit = x_decoded[0].reshape(digit_size, digit_size)
        VAE_figure[(n-i-1) * digit_size: (n - i) * digit_size,
                   j * digit_size: (j + 1) * digit_size] = normalized(digit)
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
XXX, YYY = np.meshgrid(grid_x_vae, grid_y_vae)
plt.scatter(XXX, YYY)

plt.subplot(1, 2, 2)
plt.imshow(VAE_figure, cmap='Greys')
plt.axis('off')
plt.show()
```

### 動態比較

```python
idx_1, idx_2 = np.random.randint(X_train.shape[0], size=2)
```

```python
def inBetween(t):
    data_0 = X_train[idx_1].reshape(28, 28)
    data_1 = X_train[idx_2].reshape(28, 28)
    data_t = (1-t)*data_0 + t*data_1
    
    mu_0 = VAE_Encoder.predict(X_train[idx_1:idx_1+1]).squeeze()
    mu_1 = VAE_Encoder.predict(X_train[idx_2:idx_2+1]).squeeze()
    mu_t = (1-t)*mu_0 + t*mu_1

    plt.figure(figsize=(10, 5))
    ax1 = plt.subplot(2, 1, 2)
    ax1.scatter(mu_0[0], mu_0[1])
    ax1.scatter(mu_1[0], mu_1[1])
    ax1.scatter(mu_t[0], mu_t[1])

    ax2 = plt.subplot(2, 3, 1)
    ax2.imshow(data_0, cmap='Greys')
    ax2.set_title('t=0')

    ax3 = plt.subplot(2, 3, 2)
    ax3.imshow(data_t, cmap='Greys')
    ax3.set_title(f't={t}')

    ax4 = plt.subplot(2, 3, 3)
    ax4.imshow(data_1, cmap='Greys')
    ax4.set_title('t=1')
```

```python
interact(inBetween, t=FloatSlider(value=0, 
                                  min=0, 
                                  max=1.0,
                                  step=0.02,))
```
