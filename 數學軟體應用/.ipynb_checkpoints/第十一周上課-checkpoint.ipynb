```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## PCA

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
from sklearn.decomposition import PCA
```

## Step 1.開 PCA 機

```python
pca = PCA(n_components=2)
## n_components=2 壓到二維
```

## Step 2.訓練

```python
pca.fit(x)
```

## Step 3.預測

```python
x[87]
```

```python
X = pca.transform(x)
```

```python
X[87]
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='Set3')
```

## 準備訓練/測試資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Step 1. 開 SVC 函數學習機

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma='auto')
```

## Step 2.訓練

```python
clf.fit(x_train, y_train)
```

## Step 3. 預測

```python
y_predict = clf.predict(x_test)
```

```python
y_predict
```

```python
x0 = np.arange(-4, 4.2, 0.02)
y0 = np.arange(-1.5, 1.7, 0.02)

xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, alpha=0.3, cmap='Set3')
plt.scatter(x_test[:,0], x_test[:,1], c=y_test, cmap='Set3')
```

## K-Means 自動分類

```python
x = np.random.rand(100,2)
```

```python
plt.scatter(x[:,0], x[:,1])
```

## Step 1. 開 K-Means 函數學習機

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3)
```

## Step 2.訓練

```python
clf.fit(x)
```

## Step 3. 預測

```python
clf.predict(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c=clf.labels_)
```

```python
x0 = y0 = np.arange(-0.2, 1.2, 0.02)
xm, ym = np.meshgrid(x0,y0)

P = np.c_[xm.ravel(),ym.ravel()]
z = clf.predict(P)

Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, alpha=0.3, cmap='Set3')
plt.scatter(x[:,0], x[:,1], c=clf.labels_, cmap='Set3')
```

## make_blobs生更像真實世傑的數據

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
x, y = make_blobs(n_samples=500, centers=3, n_features=2, random_state=0)
```

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap='Set3')
```

```python
plt.scatter(x[:,0], x[:,1])
```

## Step 1. 開 K-Means 函數學習機

```python
clf = KMeans(n_clusters=3)
```

## Step 2.訓練

```python
clf.fit(x)
```

## Step 3. 預測

```python
plt.scatter(x[:,0], x[:,1], c=clf.labels_, cmap='Set3')
```

## Mean-Shift 完全自動分類

#### Step 1. 開 Mean-Shift 函數學習機

```python
from sklearn.cluster import MeanShift
```

```python
clf = MeanShift()
```

#### Step 2.訓練

```python
clf.fit(x)
```

#### Step 3. 預測

```python
clf.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c=clf.labels_, cmap='Set3')
```

```python
x0 = np.arange(-4.5, 5.2, 0.02)
y0 = np.arange(-3, 7, 0.02)

xm, ym = np.meshgrid(x0, y0)

P = np.c_[xm.ravel(),ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)

plt.contourf(xm, ym, Z, alpha=0.3, cmap='Set3')
plt.scatter(x[:,0], x[:,1], c=clf.labels_, cmap='Set3')
```

```python
clf = MeanShift(bandwidth=0.2)
```

```python
clf.fit(x)
```

```python
plt.scatter(x[:,0], x[:,1], c=clf.labels_, cmap='Set3')
```

```python
x0 = np.arange(-4.5, 5.2, 0.02)
y0 = np.arange(-3, 7, 0.02)

xm, ym = np.meshgrid(x0, y0)

P = np.c_[xm.ravel(),ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)

plt.contourf(xm, ym, Z, alpha=0.3, cmap='Set3')
plt.scatter(x[:,0], x[:,1], c=clf.labels_, cmap='Set3')
```
